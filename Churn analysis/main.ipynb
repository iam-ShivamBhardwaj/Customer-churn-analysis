```python

# Assuming 'df' is your dataframe and it has the columns 'SeniorCitizen' and 'Churn'
# Create a cross-tabulation of 'SeniorCitizen' and 'Churn'
cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plotting the stacked bar chart
ax = cross_tab.plot(kind='bar', stacked=True, figsize=(4, 4))

# Adding labels and title
plt.title("SeniorCitizen vs Churn (Stacked Bar Chart with %)", fontsize=16)
plt.xlabel("Senior Citizen", fontsize=12)
plt.ylabel("% of Total", fontsize=12)

# Adding percentage labels on the bars
for bar_group in ax.containers:
    for bar in bar_group:
        height = bar.get_height()
        if height > 0:  # Only add labels for non-zero segments
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f"{height:.1f}%",
                ha="center",
                va="center",
                fontsize=10,
                color="black"
            )

plt.legend(title='Churn', bbox_to_anchor=(0.9,0.9)) # bbox_to_anchor=(0.9,0.9) isse legend() chrun yes no) box ke bahar jayega
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### Analysis on services

```python
# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Determine the grid size for subplots
n_cols = 3  # Number of columns in the grid
n_rows = -(-len(columns) // n_cols)  # Calculate rows needed

# Create a figure
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
axes = axes.flatten()  # Flatten to easily index each subplot

# Iterate through each column and create a countplot
for i, col in enumerate(columns):
    ax = axes[i]
    sns.countplot(data=df, x=col, ax=ax, hue=df['Churn'])
    ax.set_title(f"{col}", fontsize=12)
    ax.set_xlabel("")
    ax.set_ylabel("Count", fontsize=10)
    ax.tick_params(axis='x', rotation=30)

# Hide unused subplots if the grid is larger than the number of columns
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

#### Customers with no phone service or internet-related add-ons (e.g., online security, backup, or device protection) show lower churn rates compared to those with active services. Notably, fiber optic internet and streaming services exhibit higher churn rates.
